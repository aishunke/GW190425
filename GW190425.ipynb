```python
import numpy as np

def reduce_dimension(ask):
    output=[]
    for i in range(len(ask)):
        for j in range(len(ask[i])):
            output.append(ask[i][j])
    return output



def split(ask):
    output=[]
    output1=''
    
    for i in range(len(ask)):
        digit=ask[i]
        if digit ==' ' or digit=='\t': 
            if len(output1)>0:
                output.append(output1)
            output1=''
            continue
        if digit == '\n':
            output.append(output1)
            output1=''
            break
        else: 
            if i == len(ask)-1:
                output1+=digit
                output.append(output1)            
        output1+=digit
    return output

def read_row(ask):
    after_reduce=reduce_dimension(ask)
    row_Num=len(split(after_reduce[0]))   
    row=np.zeros((len(after_reduce),row_Num))
    for i in range(len(after_reduce)):
        line=split(after_reduce[i])
        for j in range(row_Num):
            row[i][j]=float(line[j]) 
    output=[]        
    for k in range(row_Num):
        output.append(row[:,k])
    return output
    

def read_row_txt(file_name):
    data_list=[]
    with open(file_name,'r') as f1:
        data_list.append(f1.readlines())
    f1.close    
    output=read_row(data_list)
    return output
```
